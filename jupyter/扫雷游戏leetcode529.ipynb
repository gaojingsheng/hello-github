## Leetcode 529
扫雷游戏

```python
class Solution(object):
    def mine_num(self,board,x,y):
        length = len(board)
        width = len(board[0])
        result = 0
        for i in range(max(x-1,0),min(x+2,length)):
            for j in range(max(y-1,0),min(y+2,width)):
                if i != x or j!= y:
                    if board[i][j] == "M":
                        result += 1
        return result
    def isclick(self,board,x,y):
        if board[x][y] == "E" or board[x][y] == "M":
            return True
        else:
            return False
    def click_once(self,board,x,y):
        if not self.isclick(board,x,y):
            return board

        length = len(board)
        width = len(board[0])
        number = 0
        if self.mine_num(board,x,y) == 0:
            board[x][y] = 'B'
        elif self.mine_num(board,x,y) > 0:
            board[x][y] = str(self.mine_num(board,x,y))
            return board

        for i in range(max(x-1,0),min(x+2,length)):
            for j in range(max(y-1,0),min(y+2,width)):
                if i != x or j!= y:
                    if board[i][j] == "M":
                        pass
                    elif self.mine_num(board,i,j) == 0 and self.isclick(board,i,j):
                        board = self.click_once(board,i,j)
                    elif self.mine_num(board,i,j) > 0 and self.isclick(board,i,j):
                        board[i][j] = str(self.mine_num(board,i,j))
        return board

    def updateBoard(self, board, click):
        """
        :type board: List[List[str]]
        :type click: List[int]
        :rtype: List[List[str]]
        """
        if not board:
            return 0
        length = len(board)
        width = len(board[0])
        x = click[0]
        y = click[1]
        if not self.isclick(board,x,y):
            print("click is not a useful button!")
            print(board)
            return board
        if board[x][y] == "M":
            print("click the mine")
            board[x][y] = "X"
            print(board)
            return board
        self.click_once(board,x,y)
        click_button = 0
        for i in range(length):
            for j in range(width):
                if self.isclick(board,i,j):
                    click_button += 1
                    break
            else:
                continue
            break
        if click_button == 0:
            print("useful button is zero")
            print(board)
            return board
            
        return board
```
